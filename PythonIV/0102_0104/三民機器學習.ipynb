# KNN

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# KNN分類器
from sklearn.neighbors import KNeighborsClassifier
# 訓練與測試分割
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
# 做預測時，通常x是題目 y是答案
# x是150題目 y是150答案
x = iris.data
y = iris.target
x,y
```

```python
# 隨機切取四等分 訓練題目 測試題目 訓練答案 測試答案
# test_size 如果是小數點為總數的百分比，如果是整數為筆數
# random_state 亂樹種子 0~42之間，1每次都固定，如果未指定為隨機
# random_state=1 讓訓練樣本固定，再將其他因子作微調，以便達到測試的目的
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# 建模
# p=2 使用阿斯里德距離
knn = KNeighborsClassifier(n_neighbors=3, p=2, weights='distance')
```

```python
# 訓練
knn.fit(X_train, y_train)
```

```python
# 預測
knn.predict(X_test)
```

```python
y_test
```

```python
# 測試分數 八九成以上已經OK
knn.score(X_test, y_test)
```

# 找K值

```python
from sklearn import metrics
```

```python
# 正確率
accuracy = list()
# 建模 找K值
def findK():
  for k in range(1, 100):
    # 把每個鄰居給他，預測準度
    knn = KNeighborsClassifier(n_neighbors=k, p=2)

    # 訓練
    knn.fit(X_train, y_train)

    # 預測
    y_pred = knn.predict(X_test)

    # 將測試與預測做正確度的比較
    accuracy.append(metrics.accuracy_score(y_test, y_pred))

  k_range = range(1, 100)
  # 畫折線圖
  plt.plot(k_range, accuracy)
  plt.show()

findK()
```

```python
#accuracy
accuracy.count(0.9777777777777777)
```

# IRIS 鳶尾花分類

```python
tree_X = iris['data']
tree_y = iris['target']
# 測試三成
X_train, X_test, y_train, y_test = train_test_split(tree_X, tree_y, test_size=0.3, random_state=1)
```

```python
# 建模
# 隨機森林決策樹
from sklearn.ensemble import RandomForestClassifier

# n_estimators 樹(數量) criterion 演算法
rfc = RandomForestClassifier(n_estimators=5, criterion='entropy')
```

```python
# 訓練
rfc.fit(X_train, y_train)
```

```python
# 預測
rfc.predict(X_test)
```

```python
# 正解
y_test
```

```python
# 測試分數
rfc.score(X_test, y_test)
```

```python
# 訓練與測試分數 --> 理論為1
rfc.score(X_train, y_train)
```

1. 樹太少會發生上下起伏的問題
2. 樹的數量太少會影響過度擬合問題
